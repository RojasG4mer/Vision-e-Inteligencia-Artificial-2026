## Procesamiento de imágenes:
Partimos de una imagen y obtener otra nueva.




#### Técnicas
- Mejorado de imágenes.
- Compresión de imágenes.
- Restauración de imágenes.
  - Imágenes fuera de foco.
- Identificación de ROI.

#### Reconocimiento de patrones
Identificar los objetos existentes en una imagen.

Partimos de una imagen 2D, ahora aplicamos:

- Segmentación
- Filtros
- Identificación de bordes
- Clasificación y reconocimiento de patrones.

Ahora obtenemos un patrón o patrones con estas técnicas.

Otras técnicas de patrones

- Reconocimiento de rostros.
- Reconocimiento de huellas digitales.
- Reconocimiento de células.
- Reconocimiento de placas.

# Dificultades de la visión por computadora

Es un mapeo de M:1 (3D-> 2D)
- Muchas superficies 3D con materiales, geometría e iluminación distintas, nos llevan a imágenes 2D idénticas.
- El mapeo inverso (2D->3D) no tiene una solución única, porque en el paso 3D -> 2D se ha perdido información.

Computacionalmente cara.
- El cerebro humano trabaja en paralelo, para procesar miles de señales. Una PC tiene un solo 

Dificultad para identificar el patrón a reconocer.
- No entendemos aun el problema de reconocimiento de patrones

### Cámaras
Es un dispositivo electrónico que captura y almacena imágenes en formato digital en lugar de usar película química. Utiliza un sensor de imagen para convertir a luz en señales electrónicas, las cuales se procesan y se guardan en tarjetas de memoria.

#### Tipos de cámaras
1. Lineales: Realizan un barrido lineal para la construcción de una imagen línea a línea, de forma que el objeto se desplaza
2. Matriciales: El sensor de la cámara cibre un área formada por una matriz de píxeles.

### Descripción de iluminación 
Su función es ayudar a captar una imagen clara del producto.

El sistema de iluminación es una parte muy crítica en el sistema

#### Tipos de iluminación:
- Iluminación posterior difusa. 
- Iluminación posterior direccional.
- Iluminación frontal oblicua.
- Iluminación frontal direccional.
