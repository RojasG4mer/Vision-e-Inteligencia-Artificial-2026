## Escala de grises
Formas de escala de grises apartir de una imagen RGB
- Average:
  $$ \frac{R+G+B}{3} $$
- Lightness:
  $$ \frac{max(R, G, B) + min(R, G, B)}{2} $$
- Luminosity: (se usa mayormente en los softwares de visión para convertir a escala de grises)
  $$ (0.21\times R) + (0.72\times B) + (G+0.11 \times B) $$
- Luminancia:
  $$ (0.3\times R) + (0.59\times G) + (0.11\times B) $$

Al tener una escala de grises, ya solo tenemos un canal, así que podemos sacar un histograma de este.

## Operaciones aritméticas y lógicas



### Suma ponderada
Fusionamos dos imágenes
$$ Y = a_1 x_1 + a_2 x_2 $$


### Suma - promedio
Fusionamos varias imágenes, que si tienen ruido, podemos reducir el ruido.
$$ Y = \frac{1}{n} \sum_{i=1}^{n}x_i $$


### Resta
Nos ayuda a identificar cosas que estén diferentes entre las dos imágenes que se usan.
$$ Y = X_1 - X_2 $$

### Mutiplicación por un escalar
Ajustar contrastes o intensidad 
$$ Y = kX_1 $$

### Mutiplicación de imágenes (producto punto)
Con esto hacemos una extracción de la zona donde tenemos un valor diferente de cero (vale 1)
$$ Y = X_1 \cdot X_2 $$
