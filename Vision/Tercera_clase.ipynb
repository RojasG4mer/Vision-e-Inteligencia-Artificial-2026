# Conteo de objetos


## Procedimiento
1. Convertirla a escala de grises
2. Convertirla a binarizarla
3. Se requiere que los objetos estén en blanco (1) y el fondo en negro (0)
4. Aplicar una técnica para identificar los objetos (depende de la técnica a usar)
5. Obtener el número de objetos.



```python
ruta = '../imagenes/Casa_grises.jpg'
ima_gray = cv2.imread(ruta, cv2.IMREAD_GRAYSCALE)
# ima_gray = cv2.imread(ruta)
print(ima_gray.shape)
binary = ima_gray < 90
ima_bin = binary.astype('uint8') * 255
plt.imshow(ima_gray, cmap='gray')
plt.show()
plt.imshow(ima_bin, cmap='gray')
plt.show()
```

```python
contar , _ = cv2.findContours(ima_bin, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print('Número de objetos:', len(contar))
```

```python
num_labels, labels = cv2.connectedComponents(ima_bin)
print('Número de objetos:', num_labels - 1)
```
